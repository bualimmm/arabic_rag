```python
from docx import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
def split_word_document(file_path):
    """Reads a Word document and splits it into chunks."""

    doc = Document(file_path)
    text = "\n".join([para.text for para in doc.paragraphs])  # Extract text

    # Customize the splitter (adjust parameters as needed)
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=100
    )

    chunks = text_splitter.split_text(text)
    return chunks

# Example Usage:
file_path = "/content/extracted_text.docx"  # Replace with your file path
chunks = split_word_document(file_path)

# Process or analyze the chunks
for i, chunk in enumerate(chunks):
    print(f"Chunk {i + 1}: {chunk}")
```

```python
import re  # Import the regular expression library

def remove_substrings(strings):
    # Regular expression to match [i] where i is an integer
    pattern = r"\[\d+\]"
    # Apply the pattern to remove matches from each string in the list
    return [re.sub(pattern, '', string) for string in strings]

cleaned_chunks= remove_substrings(chunks)
# Process or analyze the chunks
for i, chunk in enumerate(cleaned_chunks):
    print(f"Chunk {i + 1}: {chunk}")
```

```python
import torch
import pandas as pd
from sentence_transformers import SentenceTransformer

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = SentenceTransformer('intfloat/multilingual-e5-large', device=device)

embeddings = model.encode(cleaned_chunks)

data = {'sentence': cleaned_chunks, 'embedding': embeddings.tolist()}
df = pd.DataFrame(data)
df.to_csv("cleaned_document_dataset.csv", index=False)
```
